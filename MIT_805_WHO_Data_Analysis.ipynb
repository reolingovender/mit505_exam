```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr
```

```python
data_1 = pd.read_csv('/content/drive/MyDrive/MIT805_Exam/WHO_HealthData.csv')
data_2 = pd.read_csv('/content/drive/MyDrive/MIT805_Exam/WHO_HealthData_Meta.csv')
```

```python
# Rename and merge metadata and data
data_2.rename(columns={'Label': 'GHO'}, inplace=True)
merged_data = data_1.merge(data_2, on='GHO', how='inner')
```

```python
# Remove columns with 'Unnamed' in their names
merged_data = merged_data.loc[:, ~merged_data.columns.str.contains('^Unnamed')]
```

```python
# Corrected code with parentheses around each condition
filtered_data = merged_data[(merged_data['COUNTRY'] == 'ZAF') | (merged_data['REGION'] == 'GLOBAL')]
```

```python
# List of specific values to filter
values_to_select = [
    "WHOSIS_000001", "WHOSIS_000002", "HIV_0000000026", "MDG_0000000020",
    "MALARIA_EST_INCIDENCE", "NCDMORT3070", "RS_196", "SDGSUICIDE",
    "NCD_HYP_PREVALENCE_A", "NCD_BMI_30A", "HWF_0002", "HWF_0003", "HWF_0004",
    "HWF_0007", "HWF_0011", "HWF_0015", "HWF_0016", "HWF_0017", "HWF_0018",
    "HWF_0019", "HWF_0020", "HWF_0021", "HWF_0022", "HWF_0023", "HWF_0024"
]

# Filter the dataframe based on these values in the 'GHO' column
filtered_data = filtered_data[filtered_data['GHO'].isin(values_to_select)]
```
